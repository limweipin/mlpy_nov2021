In this hands-on, we apply k-means clustering techniques to visualize the dataset of truck drivers and to group them into different categories.


```python
%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbow
from warnings import filterwarnings
filterwarnings('ignore')

df = pd.read_csv('data/drivers.csv', sep='\t', header=0)
X = df.iloc[:, 1:].values
```

- To do: Use Elbow method to find the optimum number of clusters

```python
inertias = []
for k in range(1, 10):
    km = KMeans(n_clusters=k).fit(X)
    inertias.append(km.inertia_)
    
plt.plot(range(1, 10), inertias, 'b*-')
plt.xlabel('Number of cluster')
plt.ylabel('Inertia')
plt.title('Elbow method')
plt.show()
```

- To do: Apply k-mean clustering with the optimum number of clusters obtained using the Elbow method

```python
km = KMeans(n_clusters=2).fit(X)
```

```python
# Plot the resulting clusters
plt.style.use('default')
plt.scatter(X[:, 0], X[:, 1], c=km.labels_)
plt.show()
```

- To do: Apply k-mean clustering with 4 clusters and plot the resulting clusters

```python
km = KMeans(n_clusters=4).fit(X)

plt.style.use('default')
plt.scatter(X[:, 0], X[:, 1], c=km.labels_)
plt.show()
```

- To do: Save the features and labels into a new csv file

```python
df['Label'] = km.labels_
df.to_csv('data/truck_driver_labelled.csv', index=False)
```
