```python
# %load im01_read_csv.py
import csv
import pandas as pd
from collections import defaultdict

csv_fname = 'data/spx.csv'

# reading csv file using csv module
with open(csv_fname, 'r') as csv_file:
    reader = csv.reader(csv_file, delimiter=',', quotechar='"')
    header = next(reader)
    print(header)
    data = defaultdict(list)
    for row in reader:
        for key, field in zip(header,row):
            data[key].append(field)
            
# loading csv file using pandas
df = pd.read_csv(csv_fname, delimiter=',', parse_dates=['Date'])
```

```python
# %load im01_read_html.py
import requests
import bs4

# downloading a webpage and save into a html file
url = "https://beautiful-soup-4.readthedocs.io/en/latest/#quick-start"
resp = requests.get(url)
with open('bs4_doc.html', 'w', encoding='utf8') as f:
    f.write(resp.text)

# reading a html file and parse
with open('bs4_doc.html', 'r', encoding='utf8') as f:
    html = f.read()
    soup = bs4.BeautifulSoup(html, 'html.parser')
    for link in soup.find_all('a'):
        print(link.get('href'))
```

```python
# %load im01_read_image.py
fname = 'data/berries.jpg'

# Using PIL
from PIL import Image
im1 = Image.open(fname)
im1.show()

# Using matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread(fname)
plt.imshow(img)
plt.show()

# Using opencv
import cv2
im2 = cv2.imread(fname)
cv2.imshow('Image (Press any key to quit)', im2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
